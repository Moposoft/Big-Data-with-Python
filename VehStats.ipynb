```python
import pandas as pd
import matplotlib.pyplot as plt

with open("header.txt","r") as file :
    for line in file:
        line=line.strip()
        header=line.split(";")
print(header) 
```

```python
df = pd.read_csv("Tieliikenne Avoin Data 5.6.csv",encoding = "ISO-8859-1",sep=';')
df
```

```python
df.columns=header
df
```

```python
df = df[(df.vehicleClass == "M1") | (df.vehicleClass == "M1G") ]
df
```

```python
df=df.loc[:,["make","deploymentDate","mileage(km)","fuelType","co2"]]
df
```

```python
df.deploymentDate = df.deploymentDate.astype("str")
df.deploymentDate = df.deploymentDate.map(lambda x: x[:4])
df.deploymentDate = pd.to_datetime(df.deploymentDate,errors='coerce',format="%Y")
df["mileage(km)"]=pd.Series.abs(df["mileage(km)"])
df["fuelType"]=pd.to_numeric(df["fuelType"],errors='coerce')
df=df.dropna()
df
```

```python
plt.subplots_adjust(top=1.5, bottom=0.0, left=0.10, right=1.5, hspace=0.25, wspace=0.35)
plt.figure(figsize=(10,10))

#A helper function to show values in the pie chart
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
    return my_autopct

pie = df.make.value_counts()
plt.subplot(221)
pie[:5].plot.pie(autopct=make_autopct(pie[:5]),subplots=True)

cars = df[(df["mileage(km)"]) <= 1000000]
plt.subplot(222)
cars["mileage(km)"].plot.hist(bins=100)

data=[0,0,0,0,0]
data[0] = len(df[df.deploymentDate >= pd.datetime(2014,1,1)])
data[1] = len(df[(df.deploymentDate >= pd.datetime(2009,1,1)) & (df.deploymentDate < pd.datetime(2014,1,1))])
data[2] = len(df[(df.deploymentDate >= pd.datetime(2005,1,1)) & (df.deploymentDate < pd.datetime(2009,1,1))])
data[3] = len(df[(df.deploymentDate >= pd.datetime(1999,1,1)) & (df.deploymentDate < pd.datetime(2004,1,1))])
data[4] = len(df[df.deploymentDate < pd.datetime(1999,1,1)])
labels=["<=5","[5,10[","[10,15[","[15,20[",">20"]
plt.subplot(223)
plt.bar(labels,data)
plt.xlabel("Years old")

gas=df[df.fuelType == 1.0]
gas_co2=gas.co2
diesel=df[df.fuelType == 2.0]
diesel_co2=diesel.co2
frame = { 'Gasoline': gas_co2, 'Diesel': diesel_co2 }
co2=pd.DataFrame(frame)
plt.subplot(224)
co2.boxplot(column=['Gasoline','Diesel'],grid=False)
plt.savefig("vehstats.png")
```
