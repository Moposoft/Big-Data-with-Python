```python
import pandas as pd
from scipy.spatial.distance import squareform, pdist

original = pd.read_csv("lastfm-matrix-germany.csv",encoding = "ISO-8859-1",sep=',')

df = original.drop(columns = ["user"])
df["sum"] = df.sum(axis = 1, skipna = True)
temp = df.sum(axis = 0, skipna = True) 
cols = list(temp[temp<35].index)
df = (df.drop(columns = cols))
df = df[(df["sum"]) >= 10]
df = (df.drop(columns ="sum"))
result = squareform(pdist(df.T, "cosine"))

art=[]
for i,x in enumerate(result):
    x = list(zip(x, range(len(result))))
    x.sort()
    art.append((i, x[1][1], x[1][0]))
art2 = [x for x in art if (x[1], x[0], x[2]) not in art]

for x in art2:
    print(str(x[2]) +": " +df.columns[x[0]] +", " +df.columns[x[1]])

rihanna = sorted(list(zip(result[df.columns.get_loc("rihanna")], range(len(result)))))
nirvana = sorted(list(zip(result[df.columns.get_loc("nirvana")], range(len(result)))))
kanye = sorted(list(zip(result[df.columns.get_loc("kanye west")], range(len(result)))))

for x in range (1,11):
    print("Rihanna & " + df.columns[rihanna[x][1]] +" " +str(rihanna[x][0]))
    
for x in range (1,11):
    print("Nirvana & " + df.columns[nirvana[x][1]] +" " +str(nirvana[x][0]))

for x in range (1,11):
    print("Kanye West & " + df.columns[kanye[x][1]] +" " +str(kanye[x][0]))
```
