```python
from sklearn.linear_model import Perceptron
p = Perceptron()

with open("vertigo_train.txt","r") as file :
    data=[list(map(int, line.replace(' ','').strip())) for line in file]
    classes=[(x.pop(0)) for x in data]
print(data)
#print(classes)
p.fit(data, classes)
```

```python
with open("vertigo_predict.txt","r") as file :
    test=[list(map(int, line.replace(' ','').strip())) for line in file]
print(test)
result=p.predict(test)
print(result)
```

```python
with open("vertigo_answers.txt","r") as file :
    answers=[int(x.strip()) for x in file]
#print(answers)
got_correct=[x for x, y in zip(result, answers) if x == y]
#print(got_correct)
prcnt = (len(got_correct)/len(answers))*100
print("Perceptron (default): {:.3g}% correct".format(prcnt))
```

```python
values=[10, 50, 100, 250, 500, 1000]
for n in values:
    p = Perceptron(max_iter = n)
    p.fit(data, classes)
    result=p.predict(test)
    #print(result)
    got_correct=[x for x, y in zip(result, answers) if x == y]
    #print(got_correct)
    prcnt = (len(got_correct)/len(answers))*100
    print("Perceptron (max_iter = {}): {:.3g}% correct".format(n, prcnt))
```

```python
dist=[]
nearest=[]
for y in range(len(test)):
    dist=[]
    for x in range(len(data)):
        if x != y: dist.append(sum(list(map(lambda a,b: abs(a-b), test[y],data[x]))))
    nearest.append(dist.index(min(dist))+1)
print(nearest)
nearest_classes=[classes[x] for x in nearest]
#print(nearest_classes)
#print(answers)
p = Perceptron()
p.fit(data, classes)
result=p.predict(test)
got_correct=[x for x, y in zip(result, nearest_classes) if x == y]
prcnt = (len(got_correct)/len(nearest_classes))*100
print("Perceptron neighbour: {:.3g}% correct".format(prcnt))
```
